실습 1번 기온 공공 데이터를 이용하여 일교차가 가장 큰 날, 최저기온이 가장 낮은 날, 최고 기온이 가장 높은 날 출력

```python
import csv
f = open('ta_20220519091109.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)
max_temp = -999
min_temp = 999
check_temp = 0
cmax_temp = -1000
dates = ['','','']
for row in data:
    if row[-1]=='':
        row[-1]=-999
        row[-2]=999
    row[-1]=float(row[-1])
    row[-2]=float(row[-2])
    if row[-2]!=999:
        check_temp = row[-1]-row[-2]
        if cmax_temp<check_temp:
            dates[2] = row[0]
            cmax_temp = check_temp
    if min_temp>row[-2]:
        dates[1] = row[0]
        min_temp = row[-2]
    if max_temp<row[-1]:
        dates[0] = row[0]
        max_temp = row[-1]
print('일교차가 가장 큰 날')
print(dates[2]+'일')
print('최저기온이 가장 낮은 날')
print(dates[1]+'일')
print('최고기온이 가장 높은 날')
print(dates[0]+'일')
f.close()
```

실습 2번 지하철 노선별 역별 이용현황 데이터를 이용하여 노선별로 가장 이용객이 많은 3개 역과 가장 적은 3개 역 출력
이용객=승차+하차수

```python
import csv
f = open('CARD_SUBWAY_MONTH_202112.csv','r',encoding='UTF-8')
data = csv.reader(f)
header = next(data)
user={}
for row in data:
    #print(row)
    row[3] = int(row[3])
    row[4] = int(row[4])
    cnt = row[3]+row[4]
    if user.get(row[1])==None:
        #print('데이터 없음')
        user[row[1]] = {}
    if user[row[1]].get(row[2])==None:
        #print('신규')
        user[row[1]][row[2]] = 0
    user[row[1]][row[2]] += cnt
for each in user:
    each_sort = sorted(user[each].items(), key = lambda item: item[1])
    print('노선명: '+each)
    print('이용객이 가장 많은 역 순위')
    print('1. '+each_sort[-1][0]+' ('+str(each_sort[-1][1])+')')
    print('2. '+each_sort[-2][0]+' ('+str(each_sort[-2][1])+')')
    print('3. '+each_sort[-3][0]+' ('+str(each_sort[-3][1])+')')
    print('이용객이 가장 적은 역 순위')
    print('1. '+each_sort[0][0]+' ('+str(each_sort[0][1])+') ')
    print('2. '+each_sort[1][0]+' ('+str(each_sort[1][1])+') ')
    print('3. '+each_sort[2][0]+' ('+str(each_sort[2][1])+') ')
    print('')
    #print(each_sort)
f.close()
```
